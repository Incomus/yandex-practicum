# Determination of car prices
The service for selling used cars is developing an application to attract new customers where one can quickly find out the market value of their car. At our disposal is historical data: technical characteristics, configurations and prices of cars. We need to build a model to determine the cost.

The following point are of interest to our client:

- quality of prediction
- prediction speed
- learning time

**Data Description**

**Features**

- DateCrawled — date the profile was downloaded from the database
- VehicleType — car body type
- RegistrationYear — year of vehicle registration
- Gearbox — gearbox type
- Power — power (hp)
- Model — car model
- Kilometer - mileage (km)
- RegistrationMonth — month of vehicle registration
- FuelType
- Brand — car brand
- Repaired — whether the car was repaired or not
- DateCreated — form creation date
- NumberOfPictures — number of photos of the car
- PostalCode — postal code of the profile owner (user)
- LastSeen — date of last user activity

**Target feature**

Price — price in euros

**Purpose of the study**

- Load and study data, carry out preprocessing
- Build models to predict the cost of a car with and without boosting method
- Compare models by quality and time spent and choose the best

**Brief report:**

Received data from the car sales service /datasets/autos.csv.

Models with and without gradient boosting compared

The best model was selected according to quality and time costs

## Data overview and preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from numpy.random import RandomState
import lightgbm as lgb
import time
from sklearn.metrics import mean_squared_error
import seaborn as sns
from tqdm import tqdm

STATE = RandomState(1337)
```

```python
data = pd.read_csv('D:\\jupyter_clone\\yandex-practicum\\16. Numerical methods\\autos.csv')
data.info()
data.head()
```

DateCrawled, DateCreated, PostalCode, LastSeen and RegistrationMonth serve no purpose for us at the moment

Let's fill in missing categorical data as unknown

```python
try:
    data = data.drop(['DateCrawled', 'DateCreated', 'PostalCode', 'LastSeen', 'RegistrationMonth'], axis=1)
except:
    pass

for column in data.columns:
    if data[column].dtypes == 'object':
        data[column] = data[column].fillna('unknown')
data.info()
```

```python
for column in ['Price', 'RegistrationYear', 'Power', 'Kilometer']:
    data.boxplot(column=column)
    plt.show()
```

There are outliers in RegistrationYear and Power, let's exclude all 0.5% quantiles

```python
try:
    did_quantile
except:
    for column in ['RegistrationYear', 'Power']:
        data = data[((data[column] <= data[column].quantile(0.995))&
                     (data[column] >= data[column].quantile(0.005)))].reset_index(drop=True)
did_quantile = True
for column in ['Price', 'RegistrationYear', 'Power', 'Kilometer']:
    data.hist(column=column,bins=40)
    plt.show()
```

```python
(data.Power == 0).sum() / len(data.Power)*100
```

There are many cars with zero power and there isn't any reason to remove them. Firstly, this is a piece of data and it should mean something. Perhaps these are cars with a broken engine. Secondly, this data can affect the cost of the car and without taking it into account we will get an incorrect predictive model. Thirdly, there are 11% of them and without taking them into account we lose a lot of the data. Most of the cars have maximum mileage, which is not surprising for a used car sales service. Let's not exclude data from mileage and price.

```python
for i in ['RegistrationYear', 'VehicleType', 'Gearbox', 'Model', 
          'FuelType', 'Brand', 'Repaired', 'NumberOfPictures']:
    bar = data[i].value_counts().head(20)
    bar.plot.bar(title=i, rot=45)
    plt.show()
```

Data on the number of images can be discarded

```python
try:
    data = data.drop('NumberOfPictures', axis=1)
except:
    pass
```

To ensure that the dataset contained only relevant and informative data, the following columns were removed: 'DateCrawled', 'DateCreated', 'PostalCode', 'LastSeen', 'RegistrationMonth' and 'NumberOfPictures'

No numeric columns with missing values were found, and categorical columns with missing data were replaced with the label 'unknown'.

Boxplots were created for the numeric columns, particularly 'RegistrationYear' and 'Power', to visualize potential outliers. Based on boxplots, the 0.5% and 95.5% quantiles were removed to eliminate extreme values that may have adversely affected the analysis.

In addition, barplots were created for categorical columns to identify outliers and histograms were created to examine the distribution of the 'unknown' category. It was found that 'unknown' data was not dominant in any particular column. It was decided to leave cars with zero power.

## Model creation and training

```python
features = data.drop(['Price'], axis=1)
target = data['Price']




features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.2, random_state=STATE)
features_train, features_test, target_train, target_test = train_test_split(
    features_train, target_train, test_size=0.25, random_state=STATE)

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
features_train = encoder.fit_transform(features_train)
features_valid = encoder.transform(features_valid)
features_test = encoder.transform(features_test)
```

### RandomForestRegressor

```python
rf = pd.DataFrame()

est_values = [20, 30, 50, 80]
depth_values = [5, 7, 12, 20, 30]

progress_bar = tqdm(total=len(est_values) * len(depth_values))

for est in est_values:
    for depth in depth_values:
        start = time.time()
        model = RandomForestRegressor(random_state=STATE, max_depth=depth, n_estimators=est)
        model = model.fit(features_train, target_train)
        fit_time = time.time() - start
        start = time.time()
        predictions = model.predict(features_valid)
        rmse = mean_squared_error(target_valid, predictions) ** 0.5
        predict_time = time.time() - start
        rf_concat = pd.DataFrame({'rmse': [round(rmse)],
                                  'time fit': [round(fit_time, 2)],
                                  'time predict': [round(predict_time, 2)],
                                  'depth': [depth],
                                  'est': [est]})
        rf = pd.concat([rf, rf_concat], ignore_index=True)
        progress_bar.update(1)  
        progress_bar.set_description(f'Done depth {depth}, est {est}')


progress_bar.close()

rf
```

```python
(
sns.heatmap(rf.pivot(index='depth', columns='est', values='rmse'), annot=True, fmt='g')
   .set(title='rmse by depth and n_estimators, RandomForestRegressor',
        xlabel='n_estimators', 
        ylabel='depth')
)
plt.show()
```

```python
print(round((rf['time fit'].sum() + rf['time predict'].sum())/60, 2), 'minutes')
```

Weak dependence on the number of trees and strong dependence on depth, flattens out at ~20. ~7 minutes training time

### LGBMRegressor

```python
lgbmr = pd.DataFrame()

progress_bar = tqdm(total=len(est_values) * len(depth_values))

for est in est_values:
    for depth in depth_values:
        start = time.time()
        model = lgb.LGBMRegressor()
        params = {"max_depth": depth, "learning_rate" : 0.1, "num_leaves": 300, 'verbose': -1}
        lgb_train = lgb.Dataset(features_train, target_train)
        model = lgb.train(params, lgb_train, num_boost_round=est)
        fit_time = time.time()-start
        start = time.time()
        predictions = model.predict(features_valid)
        rmse = mean_squared_error(target_valid, predictions)**(0.5)
        predict_time = time.time()-start
        lgbmr_concat = pd.DataFrame({'rmse': [round(rmse)],
                               'time fit': [round(fit_time, 2)],
                               'time predict': [round(predict_time, 2)],
                               'depth': [depth],
                               'est': [est]})
        
        lgbmr = pd.concat([lgbmr, lgbmr_concat], ignore_index=True)
        progress_bar.update(1)  
        progress_bar.set_description(f'Done depth {depth}, est {est}')


progress_bar.close()

lgbmr
```

```python
(
sns.heatmap(lgbmr.pivot(index='depth', columns='est', values='rmse'), annot=True, fmt='g')
   .set(title='rmse by depth and n_estimators, LGBMRegressor',
        xlabel='n_estimators', 
        ylabel='depth')
)
plt.show()
```

```python
print(round((lgbmr['time fit'].sum() + lgbmr['time predict'].sum())), 'seconds')
```

Strong dependance on both depth (also flattens at ~20) and n_estimators (num_boost_round). ~40 times faster than RandomForestRegressor

## Model analysis

With a wider choice of hyperparameters, LGBMRegressor shows itself superior in both speed and prediction quality and thus is an obvious choice between the two models.

```python
lgbmr = lgbmr.sort_values(by='rmse').reset_index(drop=True).head(1)
lgbmr = lgbmr[['rmse', 'depth', 'est']]
lgbmr['features'] = 'valid'

model = lgb.LGBMRegressor()
params = {'num_boost_round':lgbmr['est'].iloc[0], "max_depth": lgbmr['depth'].iloc[0], "learning_rate" : 0.1, "num_leaves": 300, 'verbose': -1}
model = lgb.LGBMRegressor(**params)
model.fit(features_train, target_train)
predictions = model.predict(features_test)
rmse = mean_squared_error(target_test, predictions)**(0.5)
lgbmr_concat = pd.DataFrame({'rmse': [round(rmse)],
                               'depth': [lgbmr['depth'].iloc[0]],
                               'est': [lgbmr['est'].iloc[0]],
                               'features': ['test']})
lgbmr = pd.concat([lgbmr, lgbmr_concat], ignore_index=True)
lgbmr
```

Quality of LGBMRegressor model on test set converges with validation set

```python
(
pd.Series(model.feature_importances_, index=features.columns)
    .sort_values(ascending=False)
    .plot.barh(ylabel="Relevance")
)
plt.show()
```

The most important price parameters are power, year of registration, model and brand.
## General conclusion
In this project, we analyzed a dataset after preprocessing. We divided it into a target variable (price) and feature variables, then further split these into training, validation, and testing sets.

We used OrdinalEncoder to convert categorical data to numeric, making sure to fit it only on the training data to prevent data leakage.

We then built forecasting models with RandomForestRegressor and LGBMRegressor, testing different hyperparameters n_estimators and max_depth, which helped us assess model performance.

LGBMRegressor outperformed RandomForestRegressor but required wider hyperparameter coverage.

Best performing LGBMRegressor model (n_estimators=80, max_depth=12) had an RMSE of 1677, better than RandomForestRegressor (1732) and took significantly less time to train (10s vs. ~7min).

LGBMRegressor has shown the most impactful parameters to be power, year of registration, model and brand.
