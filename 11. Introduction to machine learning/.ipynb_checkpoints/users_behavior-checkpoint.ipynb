# Introduction to machine learning
At our disposal is data on the behavior of customers who have already switched to tariffs from the project of the course “Statistical data analysis”. We need to build a model for the classification task that will select the appropriate tariff. There is no need for data preprocessing - it has already been done.

Objective: to build a model with the highest possible accuracy value. To pass the project successfully, we need to bring the percentage of correct answers to at least 0.75. Check the accuracy on a test sample.

**Data Description**

Each object in the dataset is information about the behavior of one user for a month. Known:

- calls — number of calls,
- minutes — total duration of calls in minutes,
- messages — number of SMS messages,
- mb_used — consumed Internet traffic in MB,
- is_ultra — what tariff you used during the month

**Purpose of the study**

- Build models to classify a suitable tariff
- Find the model with the highest accuracy value
- Check the model for sanity

**Brief report:**

Retrieved data from the users_behavior.csv file.

3 types of classification models are built with different hyperparameters. The models are compared and tested for sanity. The dependences of the model accuracy on hyperparameters were checked.
## Data overview

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.dummy import DummyClassifier
import seaborn as sns
state = 1337
```

```python
data = pd.read_csv('D:\\Jupyter\\11. Introduction to machine learning\\users_behavior.csv')
data.info()
data.head()
```

## Sampling

```python
features, target = data.drop(['is_ultra'], axis=1), data['is_ultra']
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.2, random_state=state)
features_train, features_test, target_train, target_test = train_test_split(
    features_train, target_train, test_size=0.25, random_state=state)
```

## Model creation and testing

```python
set_depth = 8
set_est = 61

# adding results for various models in one dataframe
def get_result(models):
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    result = accuracy_score(target_valid,predictions_valid)
    new_model = pd.DataFrame({'result': [result],
                          'model_type': [model_type],
                          'max_depth': [depth],
                          'n_estimators': [est],
                          'samples_leaf': [samples_leaf],
                          'criterion': [criterion],
                          'splitter': [splitter]})

    models = pd.concat([models, new_model], ignore_index=True).dropna(subset=['model_type'])
    return models, result

#DecisionTreeClassifier, max_depth, min_samples_leaf
models = pd.DataFrame({'result': [0], 'model_type': [None],'max_depth': [None],'n_estimators': [None]})
model_type = 'DecisionTreeClassifier'
est = None
criterion = 'gini'
splitter = 'best'
for depth in range(1,set_depth):
    for samples_leaf in range(1,11,3):
        model = DecisionTreeClassifier(random_state=state, max_depth=depth, min_samples_leaf=samples_leaf)
        models, result = get_result(models)
decisiontree1 = models

#DecisionTreeClassifier, criterion, splitter
models = pd.DataFrame({'result': [0], 'model_type': [None],'max_depth': [None],'n_estimators': [None]})
depth = 5
samples_leaf = 4
for criterion in ['gini', 'entropy']:
    for splitter in ['best','random']:
        model = DecisionTreeClassifier(random_state=state,
                                       max_depth=5,
                                       min_samples_leaf=4,
                                       criterion=criterion,
                                       splitter=splitter)
        models, result = get_result(models)
decisiontree2 = models
decisiontree = pd.concat([decisiontree1, decisiontree2], ignore_index=True)
decisiontree_short = ( decisiontree.sort_values(by=['result'],ascending=False)
                                   .head(5)
                                   .reset_index(drop=True)
                     )

#RandomForestClassifier, n_estimators, max_depth
models = pd.DataFrame({'result': [0], 'model_type': [None],'max_depth': [None],'n_estimators': [None]})
model_type = 'RandomForestClassifier'
samples_leaf = 1
criterion = 'gini'
splitter = 'best'
for est in range(10, set_est, 10):
    for depth in range(1,set_depth):
        model = RandomForestClassifier(random_state=state, max_depth=depth, n_estimators=est)
        models, result = get_result(models)
randomforest1 = models

#RandomForestClassifier, criterion, min_samples_leaf
models = pd.DataFrame({'result': [0], 'model_type': [None],'max_depth': [None],'n_estimators': [None]})
depth = 5
n_estimators = 4
for criterion in ['gini', 'entropy']:
    for samples_leaf in range(1,11,3):
        model = RandomForestClassifier(random_state=state,
                                       max_depth=5,
                                       n_estimators=30,
                                       criterion=criterion,
                                       min_samples_leaf=samples_leaf)
        models, result = get_result(models)
randomforest2 = models
randomforest = pd.concat([randomforest1, randomforest2], ignore_index=True)
randomforest_short = ( randomforest.sort_values(by=['result'],ascending=False)
                                   .head(5)
                                   .reset_index(drop=True)
                     )

#LogisticRegression
models = pd.DataFrame({'result': [0], 'model_type': [None],'max_depth': [None],'n_estimators': [None]})
model_type = 'LogisticRegression'
samples_leaf = None
depth = None
criterion = None
est = None
model = LogisticRegression(random_state=state)
logistic, result = get_result(models)
logistic_short = ( logistic.sort_values(by=['result'],ascending=False)
                                 .head(5)
                                 .reset_index(drop=True)
                     )

models = pd.concat([decisiontree_short, randomforest_short, logistic_short], ignore_index=True)
models = models.dropna(subset=['model_type']).sort_values(by=['result'], ascending=False).reset_index(drop=True)
models
```

The best model turned out to be RandomForestClassifier with a depth of 7 and the number of trees 60. The highest accuracy was shown by RandomForestClassifier, the lowest: LogisticRegression. DecisionTreeClassifier also showed high accuracy

```python
(
sns.heatmap(decisiontree1.pivot_table(index='max_depth', columns='samples_leaf', values='result'), annot=True)
   .set_title('Accuracy from depth and min_samples_leaf for DecisionTreeClassifier')
);
```

Accuracy grows up to depth 3, there is no pronounced dependence on min_samples_leaf

```python
(
sns.heatmap(decisiontree2.pivot_table(index='criterion', columns='splitter', values='result'), annot=True)
   .set_title('Accuracy from criterion and splitter for DecisionTreeClassifier')
);
```

Random splitter is worse, there is no dependence on the criterion

```python
(
sns.heatmap(randomforest1.pivot_table(index='max_depth', columns='n_estimators', values='result'), annot=True)
   .set_title('Accuracy from max_depth and n_estimators for RandomForestClassifier')
);
```

In a similar way, accuracy increases up to depth 3 in RandomForest, after which a weak dependence on n_estimators begins to appear.

```python
(
sns.heatmap(randomforest2.pivot_table(index='criterion', columns='samples_leaf', values='result'), annot=True)
   .set_title('Accuracy from criterion and samples_leaf for RandomForestClassifier')
);
```

There is no dependence either on the criterion or on min_samples_leaf

Let's check our model on a test sample

```python
if models.loc[0,'model_type'] == 'RandomForestClassifier':
    model = RandomForestClassifier(random_state=state,
                               max_depth=models.loc[0,'max_depth'],
                               n_estimators=models.loc[0,'n_estimators'])
else:
    model = DecisionTreeClassifier(random_state=state,
                               max_depth=models.loc[0,'max_depth'],
                               min_samples_leaf=models.loc[0,'samples_leaf'])
model.fit(features_train, target_train)
predictions_test = model.predict(features_test)
result = accuracy_score(target_test,predictions_test)
print('Test sample result:',result)
print('Validation sample result:',models['result'].max())
```

The test result is consistent with the validation result

## Sanity test

```python
model = DummyClassifier()
models, result = get_result(models)
result
```

The accuracy of the best model is higher than DummyClassifier, the model is adequate. The standard strategy of DummyClassifier selects for all answers the most frequently occurring one in the source data, which is more relevant in our case.
## General conclusion

```python
models['result'].max()/result
```

The resulting model predicts on average 17% better than the null model and can be used to make more informed proposals for users.
