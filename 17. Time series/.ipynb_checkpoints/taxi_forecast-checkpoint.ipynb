# Taxi order forecasting
A taxi company collected historical data on taxi bookings at airports. To attract more drivers during peak periods, we were tasked with predicting the number of taxi orders for the next hour by building a model.

The RMSE metric value on the test sample should be no more than 48.

**Purpose of the study**

- Load the data and resample it one hour at a time
- Analyze the data
- Train different models with different hyperparameters. Make a test sample of 10% of the original data
- Check the data on a test sample and draw conclusions
## Data overview and preprocessing

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error
import seaborn as sns
from numpy.random import RandomState
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
import matplotlib.dates as mdates
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder
STATE = 1337
```

```python
data_init = pd.read_csv('D:\\jupyter_clone\\yandex-practicum\\17. Time series\\taxi.csv', index_col=[0], parse_dates=[0])
data_init.sort_index(inplace=True)
data_init = data_init.resample('1H').sum()
data_init.info()
```

Data was received and resampling was carried out for one hour at a time

```python
data_init.plot(ylabel='Taxi orders',xlabel='Date', legend=False);
```

A characteristic increase in the number of orders from spring to summer is visible.
## Model creation and training
Adding calendar signs: month, day, hour and day of the week, as well as lagging values and a moving average.
<div class="alert alert-warning">
<b>Comment</b>

Tried to add 'lag' and 'rolling_mean' to the pipeline, but I'm capitulating
</div>

```python
data_0 = data_init.copy()
data_0, data_test = train_test_split(data_0, shuffle=False, test_size=1/10)
def make_features(data_0, max_lag, rolling_mean_size):
    data = data_0.copy()
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data[f'lag_{lag}'] = data[data.columns[0]].shift(lag)

    data['rolling_mean'] = data[data.columns[0]].shift().rolling(rolling_mean_size).mean()
    data = data.dropna()
    features = data.drop(data.columns[0], axis=1)
    target = data[data.columns[0]]
    
    return features, target

def train(data_0, model, heatmap_df, lags, roll):
    features, target = make_features(data_0, lags, roll)
    features_train, features_valid, target_train, target_valid = train_test_split(features, 
                                                                                target, shuffle=False, test_size=1/9)
    model = model.fit(features_train, target_train)
    predictions = model.predict(features_valid)
    new = pd.DataFrame({'lags': [lags],
                        'roll': [roll],
                        'RMSE': [round(mean_squared_error(target_valid, predictions, squared=False), 2)]})
    heatmap_df = pd.concat([heatmap_df, new], ignore_index=True)
    return  heatmap_df
```

Comparing quality of a linear regression model and a random forest based on changes in features

```python
heatmap = pd.DataFrame()
model = LinearRegression()
for lags in range(1,5):
    for roll in range(1,5):
        heatmap = train(data_0, model, heatmap, lags, roll)
(
sns.heatmap(heatmap.pivot(index='lags', columns='roll', values='RMSE'), fmt='g')
   .set(title=f'RMSE by lags and roll for {model}',
        xlabel='roll', 
        ylabel='lags')
);
plt.show()
heatmap = pd.DataFrame()
model = RandomForestRegressor(random_state=STATE, max_depth=10, n_estimators=100)
for lags in range(1,5):
    for roll in range(1,5):
        heatmap = train(data_0, model, heatmap, lags, roll).sort_values(by='RMSE').reset_index(drop=True)
(
sns.heatmap(heatmap.pivot(index='lags', columns='roll', values='RMSE'), fmt='g')
   .set(title=f'RMSE by lags and roll for {model}',
        xlabel='roll', 
        ylabel='lags')
);
plt.show()
```

The quality of RandomForestRegressor is noticeably better than LinearRegression

Let's look at various hyperparameters for RandomForestRegressor

```python
lags = int(heatmap.lags[0])
roll = int(heatmap.roll[0])
features, target = make_features(data_0, lags, roll)
model = RandomForestRegressor(random_state=STATE)
param_grid = {
    'max_depth': [10, 15, 20], 
    'n_estimators': [100,150,200] 
}
tscv = TimeSeriesSplit(n_splits=9, gap=50)
grid_search = GridSearchCV(model, param_grid, cv=tscv, scoring='neg_root_mean_squared_error', verbose=2)
grid_search.fit(features, target)
best_depth = grid_search.best_params_['max_depth']
best_est = grid_search.best_params_['n_estimators']
```

Now let's take the best hyperparameters obtained and retest for different lag and roll

```python
heatmap = pd.DataFrame()
model = RandomForestRegressor(random_state=STATE, 
                              max_depth=best_depth, 
                              n_estimators=best_est)
for lags in range(1,5):
    for roll in range(1,5):
        heatmap = train(data_0, model, heatmap, lags, roll).sort_values(by='RMSE').reset_index(drop=True)
```

```python
(
sns.heatmap(heatmap.pivot(index='lags', columns='roll', values='RMSE'), fmt='g')
   .set(title=f'RMSE от lags и roll для {model}',
        xlabel='roll', 
        ylabel='lags')
)
plt.show()
```

Let's take the best RandomForestRegressor model and check it on a test sample

```python
model_rf = RandomForestRegressor(random_state=STATE, 
                              max_depth=best_depth, 
                              n_estimators=best_est)
try:
    lags = int(heatmap.lags[0])
    roll = int(heatmap.roll[0])
except:
    pass

heatmap = pd.DataFrame()
heatmap = train(data_0, model_rf, heatmap, lags, roll)
print(f'RMSE={heatmap.RMSE[0]} for valid; depth={best_depth}, n_est={best_est}, lags={lags}, roll={roll}')
features_train, target_train = make_features(data_0, lags, roll)
features_test, target_test = make_features(data_test, lags, roll)
model = model.fit(features_train, target_train)
predictions = model.predict(features_test)
print(f'RMSE={round(mean_squared_error(target_test, predictions, squared=False),2)} for test')
```

Training the best RandomForestRegressor model showed an rmse of 32.41 on the validation set

Test rmse at 47.49 is less than 48

```python
(
pd.Series(grid_search.best_estimator_.feature_importances_, index=features.columns)
    .sort_values(ascending=False)
    .plot.barh(title='Feature importance', ylabel="Relevancy")
)
plt.show()
```

The most important parameters were the current hour and the previous value of the number of orders.

```python
df = pd.DataFrame({'predictions': predictions, 'target_test': target_test})['2018-08-15 03:00:00':'2018-08-19 03:00:00']

fig, axs = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
date_fmt = mdates.DateFormatter('%H:%M\n%d-%b')

for i in range(3):
    axs[i].xaxis.set_major_locator(mdates.HourLocator(interval=12))
    axs[i].xaxis.set_major_formatter(date_fmt)


axs[0].fill_between(df.index, df['predictions'], color='blue', alpha=0.3)
axs[0].set_ylabel('Number of orders')
axs[0].set_title('Predictions')


axs[1].fill_between(df.index, df['target_test'], color='green', alpha=0.3)
axs[1].set_title('Target')


axs[2].fill_between(df.index, df['predictions'], color='blue', alpha=0.3, label='predictions')
axs[2].fill_between(df.index, df['target_test'], color='green', alpha=0.3, label='target')
axs[2].legend(['Predictions', 'Target'])
axs[2].set_title('Together')

plt.subplots_adjust(hspace=0.3)

plt.show()
```

The convergence of predictions and target features is clearly visible
## General conclusion
Models were trained to predict future taxi orders at airports. Calendar data such as month, day, hour and day of the week was added to the data set. Lagging values and moving average are also included as additional features. The performance of two different models, linear regression and random forest, is compared using different combinations of lag and moving average values. The random forest model demonstrates superior performance compared to linear regression at different combinations of lag and roll, with the best hyperparameters being depth of 15 and 100 trees, resulting in an RMSE metric of 47.49
